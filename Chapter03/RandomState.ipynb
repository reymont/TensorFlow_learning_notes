```python
import numpy as np

# 设置seed值，生成ndarray对象
SEED = 23455

# 基于seed产生随机数
rdm = np.random.mtrand.RandomState(SEED)

# rand函数产生随机数，返回32行2列矩阵
# 32行代表32组数据，2代表输入数据的2个特征
X = rdm.rand(32, 2)

# 为每组数据制备标签
# 若（体积+重量） ≥ 1，则 Y 赋值 0
# 若（体积+重量） < 1，则 Y 赋值 1
# Y 值为输入数据集的标签（正确答案），这里记为 Y_
Y_ = [[int(x0 + x1 < 1)] for (x0, x1) in X]

# 在神经网络中，print函数是可不写的
# 此处仅为了便于观察过程数据
print("X:\n",X)
print("Y_:\n",Y_)
```
